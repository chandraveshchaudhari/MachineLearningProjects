```python
import os
import xml.etree.ElementTree as ET
imagelist=[]
valuelist=[]


for i in full:
    #print("\n")
    tree = ET.parse(i)
  
    root = tree.getroot()
    #print(i)
    firstname = i.strip('.xml')
    filename= "/home/chaudharyubuntu/Desktop/project/tfrecord/how_to_convert_csv_to_tfrecord/txtfiles/" + firstname+".txt"
    #print(filename)
    size= tree.findall('size')
    #print(size)
    
    for y in root.findall('size'):
        
    
    
        width= y.find('width').text
        height= y.find('height').text
    
        #print(width,height)
    
    
    for x in root.findall('object'):
        #print(x)
        
        name = x.find('name')
        #width= x.find('width')
        #height= x.find('height')
        #print(name.text)
        if name.text == "Scratch_or_spot":
            #print("yes")
            for y in x.findall('bndbox'):
                #print(y)
                xmin = int(y.find('xmin').text)
                ymin = int(y.find('ymin').text)
                xmax = int(y.find('ymax').text)
                ymax = int(y.find('ymax').text)
                
            #print(filename)
            
            file = open(filename ,"a")
            #print(str("0")+" " + str(xmin) +" " + str(ymin) +" " + str(xmax)+" " + str(ymax)+" ")
            #print(name.text + " " + str(xcentre) + " " + str(ycentre) + " " + str(width) + " " + str(height)+" "+ str(ymax) )
            print(root.find('filename').text+ "," +width+ "," +height+ "," + name.text + "," + str(xmin) +"," + str(ymin) +"," + str(xmax)+ "," + str(ymax))
            #file.write(root.find('filename').text+ "," +width+ "," +height+ "," + name.text + "," + str(xmin) +"," + str(ymin) +"," + str(xmax)+ "," + str(ymax)+"\n")
            imagelist.append(root.find('filename').text)
            valuelist.extend((root.find('filename').text,width,height, name.text , str(xmin) , str(ymin) , str(xmax), str(ymax)))
        if name.text == "Dent":
            #print("yes")
            for y in x.findall('bndbox'):
                #print(y)
                xmin = int(y.find('xmin').text)
                ymin = int(y.find('ymin').text)
                xmax = int(y.find('ymax').text)
                ymax = int(y.find('ymax').text)
                
            #print(filename)
            
            file = open(filename ,"a")
            print(root.find('filename').text+ "," +width+ "," +height+ "," + name.text + "," + str(xmin) +"," + str(ymin) +"," + str(xmax)+ "," + str(ymax))
            #file.write(root.find('filename').text+ "," +width+ "," +height+ "," + name.text + "," + str(xmin) +"," + str(ymin) +"," + str(xmax)+ "," + str(ymax)+"\n")
            imagelist.append(root.find('filename').text)
            valuelist.extend((root.find('filename').text,width,height, name.text , str(xmin) , str(ymin) , str(xmax), str(ymax)))
            #filename	width	height	class	xmin	ymin	xmax	ymax
```
