## Reviews rate (1) word cloud

```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS


stopwords = set(STOPWORDS)
data = pd.read_csv("/home/chaudharyubuntu/Desktop/word_clouds_webScraping/one.csv")

wordcloud = WordCloud(  background_color='black',
                          stopwords=stopwords,
                          max_words=500,
                          max_font_size=80, 
                          random_state=42
                         ).generate(str(data))

fig = plt.figure(1)
plt.figure(figsize = (100,100))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
